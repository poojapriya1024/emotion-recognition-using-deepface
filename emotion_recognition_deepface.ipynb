```python
#import the required modules
from google.colab import files
from IPython.display import Image
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
uploaded = files.upload()
```

```python
# read image
img = cv2.imread('emotion.jpg')

# call imshow() using plt object
plt.imshow(img[:, :, : : -1])

# display that image
plt.show()
```

```python
# storing the result
result = DeepFace.analyze(img,
						actions = ['emotion'])

#printing only the emotions
emotions = result[0]['emotion']
# Sort emotions by their values in descending order
sorted_emotions = sorted(emotions.items(), key=lambda item: item[1], reverse=True)

# Get the top three emotions
top_three_emotions = sorted_emotions[:3]

#printing the emotion
print(top_three_emotions[0][0])
```
